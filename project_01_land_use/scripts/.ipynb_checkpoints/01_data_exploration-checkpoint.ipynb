## Initial data inspection

### The dataset was sucessfully loaded and inspected using geopandas. 
### Initial inspection shows polygon geometries representing county boundaries.
### The dataset includes attribute fields for geographic identifiers and names.

```python
gdf.columns
```

```python
# Attribute cleaning: column names were stripped of empty spaces and standardized to lowercase characters
gdf.columns = (gdf.columns
               .str.lower()
               .str.strip()
               .str.replace(" ","_")
              )
gdf.columns
```

```python
# removed non-essential fields msa1990, 2000, and 2010 from attribute table to somplify data
cleaned_gdf = ['gid', 'cmptrl_nbr', 'cnty_nm', 'dist_nm', 'dps_nbr', 'cnty_fips',
       'cnty_nbr', 'dist_nbr', 'grid_op','create_dt', 'create_nm', 'edit_dt', 'edit_nm', 'geometry']
```

```python
gdf = gdf[cleaned_gdf]
gdf.info()
```

```python
gdf.crs
```

## Coordinate Reference System

### The dataset is already projected in EPSG:3857 (Web Mercator). This CRS was retained to maintain consistency with the source data and to support distance-based analysis.


```python
cleaned_path = Path(r"C:\Users\Alych\Documents\GIS Portfolio\project_01_land_use\data\tx-boundaries_48_bnd\counties\txdot_counties_cleaned")
gdf.to_file(cleaned_path, index = False)
```
